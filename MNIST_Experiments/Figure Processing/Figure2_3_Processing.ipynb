```python
import numpy as np
import matplotlib.pyplot as plt
ax = df.plot(kind='bar', width=0.8, figsize=(8, 3), fontsize=11, color=['g', 'magenta', 'c', 'b', 'orange']) # x= 'Experiment'
ax.grid(axis="y", linestyle=':')
ax.set_ylabel('mean reconstruction loss \n of normal test data', fontsize = 11)
ax.set_ylim(0, 0.11)
# ax.set_xlabel('Experiment', fontsize = 11)
list2 = list(df.Experiment)
ax.set_xticklabels(list2, rotation=55, ha='right')
ax.legend(bbox_to_anchor=(1, 1), fontsize = 10)
plt.show()
```
